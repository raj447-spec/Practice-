```python
mower_df = dmba.load_data('RidingMowers.csv')
```

```python
mower_df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Separate features and target
X = mower_df.drop(columns=['Ownership'])
y = mower_df['Ownership']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)

# Train decision tree (max_depth=1)
classTreeRO92 = DecisionTreeClassifier(random_state=0, max_depth=1)
classTreeRO92.fit(X_train, y_train)

# Display classes
print("Classes: {}".format(', '.join(classTreeRO92.classes_)))

# Plot decision tree
plotDecisionTree(classTreeRO92, feature_names=X.columns[:2], class_names=classTreeRO92.classes_)

# Make predictions
y_pred = classTreeRO92.predict(X_test)
y_prob = classTreeRO92.predict_proba(X_test)

# Show predictions
print("Predicted labels:\n", y_pred)
print("Predicted probabilities:\n", y_prob)

# Evaluation
print("Accuracy on test data:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
